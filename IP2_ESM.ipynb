```python
X=np.linspace(0,100,100)
X_sqd=X**2
X_ex=np.exp(X)
```

```python
fig,ax1 = plt.subplots()
ax1.plot(X, X_sqd, 
         color='green')
ax1.set_xlabel('X',
              color="green")
ax1.set_ylabel('X^2',
               color='green')
ax1.tick_params(axis='y',
                labelcolor="green")
plt.title("Plot of X versus X^2",
         color="green")
plt.show()
```

```python
fig,ax2 = plt.subplots()
ax2.scatter(X,X_ex,
            color='blue',
            label='X versus exp(X)')
ax2.set_ylabel('exp(X)', 
               color='Blue')
ax2.tick_params(axis='y', 
                labelcolor='Blue')

plt.title('Plotting X vs exp(X)',
         color='blue')
plt.show()
```

```python
car_sales=pd.read_csv("car-sales.csv")
```

```python
car_sales.head()
```

```python
car_sales.tail()
```

```python
car_sales.info()
```

```python
car_sales['Price'] = car_sales['Price'].str.replace(r'\$|,', '', regex=True).astype('int64')
```

```python
car_sales.info()
```

```python
car_sales.head()
```

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
```

```python
car_sales.info()
```

```python
periods=len(car_sales)
start_date = pd.datetime.today().date()
sale_dates = pd.date_range(start=start_date,periods=periods)
car_sales["Sale Date"]=sale_dates
car_sales.head()
```

```python
plt.plot(car_sales['Price'],
         car_sales['Total Sales'])

plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs total sales line plot')
plt.show()
```

```python
plt.scatter(car_sales['Odometer (KM)'],
            car_sales['Price'])

plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer (KM) vs Price Scatter Plot')

plt.show()
```

```python
plt.bar(car_sales['Make'],
        car_sales['Odometer (KM)'])

plt.xlabel('Make')
plt.ylabel('Odometer(KM)')
plt.title('Make vs Odometer Bar Plot')

plt.show()
```

```python
plt.hist(car_sales["Make"])

plt.xlabel("Make")
plt.title("Histogram of Car Make")

plt.show()
```

```python
fig=plt.figure(figsize=(16,8))
ax_1=fig.add_subplot(231)
ax_1.plot(car_sales['Price'],
         car_sales['Total Sales'],
        color="blue")
ax_1.set_title('Price vs Total Sales Line Plot')


ax_2=fig.add_subplot(232)
ax_2.scatter(car_sales['Odometer (KM)'],
            car_sales['Price'],
        color="blue")
ax_2.set_title("Odometer(KM) vs Price Scatter Plot")


ax_3=fig.add_subplot(234)
ax_3.bar(car_sales['Make'],
        car_sales['Odometer (KM)'],
        color="blue")
ax_3.set_title("Make vs Odometer Bar Plot")


ax_4=fig.add_subplot(235)
ax_4.hist(car_sales['Make'],
        color="blue")
ax_4.set_title("Histogram of Car Make")


plt.show()
```

```python
sns.boxplot(x="Colour",
           y="Total Sales",
           data=car_sales)
plt.show()
```

```python
sns.lmplot(x='Total Sales',
           y='Odometer (KM)', 
           hue='Make',
           ci=None, 
           data=car_sales)

plt.xlabel("Total Sales")
plt.ylabel("Odometer(KM)")

plt.show()
```

```python
sns.pairplot(car_sales[['Odometer (KM)',
                      'Doors',
                       'Price',
                       'Total Sales']])
plt.show()
```

```python
corr=car_sales[['Total Sales',
                'Price',
                'Odometer (KM)']].corr()
print(corr)
sns.heatmap(corr,
            annot=True)
plt.show()
```

#### there is a strong negative correlation between price and Odometer, thus the higher the price the less distance the car has traveled.
#### there is a very weak positive correlation between price and total sales which likely means price has little effect on how much a given car model is selling due to consumers having different price points
#### finally there is a relatively weak, negative correlation between total sales and the odometer, this likely points to a higher mileage(likely older and more used) being less sellable overall.
